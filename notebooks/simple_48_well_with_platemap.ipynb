```python
from dotenv import load_dotenv
load_dotenv()

from functools import reduce
from pathlib import Path

import pandas as pd
from platechain import parse_plates
```

```python
byonoy_df = pd.read_csv(Path.cwd().parent / 'examples/dummy_48_well_example.csv')
# Dataframe contains both results and platemap information
byonoy_df
```

```python
# Can optionally add number of plates, number of rows, and number of columns here. 
res = parse_plates(byonoy_df)
res
```

```python
# Once you have the plates as tidy dataframes, you can merge them to create a clean dataset
merged = reduce(
        lambda left, right: pd.merge(left, right, on=['row', 'column', 'well'], how="outer"),
        res,
    )
merged
```
